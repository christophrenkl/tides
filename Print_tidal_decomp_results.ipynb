```python
import pandas as pd
import os
import glob
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    print()
```
